```python
risk_of_this_career = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, "Risk of this career", risk_of_this_career)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Risk of this career')
```

```python
low_risk[['Undergraduate Major', 'Risk of this career']].head()
```

```python
high_potential_career = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending = False)
```

```python
spread_in_salaries = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```
